```python
modData = df.append({'day' : np.nan , 'month': 3.6, 'year':0, 'cases':0.2, 'deaths':0 } , ignore_index=True)
modData.describe()
```

```python
print('Columns with missing values')
print(modData.isnull().sum())
print('\n Columns with zero values')
print((modData[['day','month','year','cases','deaths']]==0).sum())
```

```python
modData[['day',
         'month',
         'year',
         'cases','deaths']] = modData[['day',
                                              'month',
                                              'year',
                                              'cases',
                                              'deaths']].replace(0, np.NaN)
print('Columns with missing values')
print(modData.isnull().sum())
```

```python
modData.fillna(modData.mean(), inplace=True)
print(modData.isnull().sum())
```

```python
X= df['cases']
Y= df['deaths']
df2 = pd.DataFrame({'cases': X, 'deaths': Y})
print(df2)
```

```python
plt.scatter(df2[['cases']], df2[['deaths']], color = "r", marker = "x", s = 15)
plt.xlabel('cases') 
plt.ylabel('deaths')
plt.title('Scatter Plot') 
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

classifier = LinearRegression()

model = classifier.fit(df2[['cases']], df2[['deaths']])
```

```python
y_pred = classifier.predict(df2[['cases']])
print(y_pred)
```

```python
plt.scatter(df2[['cases']], df2[['deaths']], color = "m", marker = "s", s = 10) 

plt.plot(df2['cases'], y_pred, color = "g") 
plt.xlabel('cases') 
plt.ylabel('deaths')
plt.title('Regression Function')
plt.show()
print('Coefficients: \n', classifier.coef_)
print('Intercept: \n', classifier.intercept_)
```

```python
Day_10537 = classifier.intercept_+ classifier.coef_ * 10537 
print(Day_10537)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error

print("Mean squared error:")
print(mean_squared_error(df2[['deaths']], y_pred))
```

```python
plt.scatter(y_pred, (df2[['deaths']] - y_pred) ** 2, color = "blue", s = 10,) 
plt.title("Squared errors")

plt.hlines(y = 0, xmin = 0, xmax = 140, linewidth = 2) 
plt.xlabel('predicted values')
plt.ylabel('squared error')
plt.show()
```

```python
print("Mean absolute error: ")
print(mean_absolute_error(df2[['deaths']], y_pred))

#plot
plt.scatter(y_pred, (df2[['deaths']] - y_pred), color = "blue", s = 10,) 
plt.title("Absolute Errors")
plt.hlines(y = 0, xmin = 0, xmax = 140, linewidth = 2) 
plt.xlabel('predicted values')
plt.ylabel('absolute error')
plt.show()
```
